```python
import pandas as pd
import numpy as np
```

```python
movie_data = pd.read_csv("Resources/Blockbusters_2019_1977.csv", encoding='latin-1')
movie_data.head()
```

```python
movie_data.info()
```

```python
movie_data[['film_budget','worldwide_gross','domestic_gross']] = movie_data[
    ['film_budget', 'worldwide_gross', 'domestic_gross']]\
        .apply(lambda x: x.str.replace(',', ''))
movie_data.head()
```

```python
movie_data['film_budget'] = movie_data['film_budget'].astype(int)
movie_data['worldwide_gross'] = movie_data['worldwide_gross'].astype(int)
movie_data['domestic_gross'] = movie_data['domestic_gross'].astype(int)
movie_data.head()
```

```python
movie_data.replace('', np.nan, inplace=True)
movie_data.head()
```

```python
movie_data[['mpaa_rating','film_title','domestic_distributor','genre_1','genre_2','genre_3']] = movie_data[[
    'mpaa_rating','film_title','domestic_distributor','genre_1','genre_2','genre_3']].apply(lambda x: x.str.strip())
movie_data.head()
```

```python
movie_data['genre_2'] = movie_data['genre_2'].replace({'Music':'Musical','romance':'Romance'})
movie_data['genre_3'] = movie_data['genre_3'].replace({'Music':'Musical'})
```
